```python
from alpha_vantage.timeseries import TimeSeries
from alpha_vantage.techindicators import TechIndicators
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time

#my api key for alpha_vantage
key = '#############'
```

```python
#it threw an error code after 83 becasue the ticker symbol was not correct 
#restarted it so that I could get the rest of the data
ts = TimeSeries(key, output_format='pandas')
i =84 #this let me keep track of where the loop was at in the process, it took probably 2 hours to run
stock_df = pd.DataFrame() #an empty dataframe I can append my data to
for stock in constituent_list[83:]: #this for loop goes through and gets stock data for the individual stocks in the S&P 500
    new_data, new_meta_data = ts.get_daily(symbol=stock, outputsize='full') #the api i am using
    new_data['symbol'] = stock 
    new_data = new_data.reset_index()
    new_data.columns = ['date', 'open', 'high', 'low', 'close', 'volume', 'symbol']
    stock_df = stock_df.append(new_data)
    print(i)
    i += 1
    time.sleep(15) #had to use this because the free edition of the api only allows 5 calls per minute
```
